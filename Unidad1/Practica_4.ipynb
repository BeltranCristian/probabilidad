**Practica 1: Diagramas de cajas y bigotes**

```python
import random
import matplotlib.pyplot as ptl
from statistics import *
Fabrica1 = []
for i in range(100):
  numero_aleatorio = round( random.uniform(50,100), 2 )
  Fabrica1.append(numero_aleatorio)
print(Fabrica1)
```

```python
Fabrica2 = []
for i in range(100):
  numero_aleatorio = round( random.uniform(40,93), 2 )
  Fabrica2.append(numero_aleatorio)
print(Fabrica2)
```

```python
ptl.boxplot(Fabrica1)
ptl.ylabel('Produccion en mts.')
ptl.title('Produccion en Fabrica 1')
ptl.show()
```

```python
ptl.boxplot(Fabrica2)
ptl.ylabel('Produccion en mts.')
ptl.title('Produccion en Fabrica 2')
ptl.show()
```

```python
fig, (ax1, ax2) = ptl.subplots(nrows=1, ncols=2, figsize=(9,4))
bplot1 = ax1.boxplot(Fabrica1,
                      vert=True,
                      patch_artist=True)
ax1.set_title('Produccion en Fabrica 1')

bplot2 = ax2.boxplot(Fabrica2,
                      vert=True,
                      patch_artist=True)
ax2.set_title('Produccion en Fabrica 2')
ptl.show()
```

```python
fig, (ax1, ax2) = ptl.subplots(nrows=1, ncols=2, figsize=(9,5))
bplot1 = ax1.boxplot(Fabrica1,
                      vert=True,
                      patch_artist=True)
ax1.set_title('Produccion en Fabrica 1')
ax1.set_ylim(0,160)

bplot2 = ax2.boxplot(Fabrica2,
                      vert=True,
                      patch_artist=True)
ax2.set_title('Produccion en Fabrica 2')
ax2.set_ylim(0,160)
ptl.show()
```
