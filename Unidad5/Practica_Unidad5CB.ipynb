```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
```

## Función para analizar cada grupo

```python
def analizar_grupo(ruta):
    print(f"\n===== ANALIZANDO {ruta} =====\n")

    df = pd.read_csv(ruta)
    display(df.head())
    display(df.describe())

    # Histograma
    plt.figure()
    plt.hist(df['Calificacion'])
    plt.title(f"Histograma - {ruta}")
    plt.xlabel("Calificación")
    plt.ylabel("Frecuencia")
    plt.show()

    # Variables correctas
    X = df[['HorasEstudio']]
    y = df['Calificacion']

    # Modelo de regresión
    model = LinearRegression().fit(X, y)

    print("\nEcuación de regresión:")
    print(f"Calificación = {model.coef_[0]:.4f} * HorasEstudio + {model.intercept_:.4f}")

    # Correlación
    r, p = pearsonr(df['HorasEstudio'], df['Calificacion'])
    print(f"\nCoeficiente de correlación r = {r:.4f}")
    print(f"p-valor = {p:.6f}")

    # Gráfica de dispersión + regresión
    plt.figure()
    plt.scatter(df['HorasEstudio'], df['Calificacion'])
    plt.plot(df['HorasEstudio'], model.predict(X))
    plt.title(f"Regresión Lineal - {ruta}")
    plt.xlabel("Horas de Estudio")
    plt.ylabel("Calificación")
    plt.show()

    # Interpretación
    print("\nInterpretación:")
    if r > 0:
        print("• Relación positiva: más horas → mayor calificación.")
    else:
        print("• Relación negativa.")

    if abs(r) > 0.7:
        print("• Relación fuerte.")
    elif abs(r) > 0.3:
        print("• Relación moderada.")
    else:
        print("• Relación débil.")

    if p < 0.05:
        print("• El modelo es estadísticamente significativo.")
    else:
        print("• El modelo NO es estadísticamente significativo.")
```

## Ejecutar análisis (archivos ya subidos a /content/)

```python
analizar_grupo('/content/grupo1.csv')
analizar_grupo('/content/grupo2.csv')
analizar_grupo('/content/grupo3.csv')
```
